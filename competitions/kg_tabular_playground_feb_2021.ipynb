```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from fastai.tabular.all import *
```

```
train_path = '/content/train.csv'
test_path = '/content/test.csv'
```

```
df = train_df
```

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df.iloc[:,1:11] = df.iloc[:,1:11].apply(le.fit_transform)
```

```
df = df.drop("id", axis=1)
```

```
from scipy.stats import pearsonr
def corr_pvalue(df):
    numeric_df = df.dropna()._get_numeric_data()
    cols = numeric_df.columns
    mat = numeric_df.values

    arr = np.zeros((len(cols),len(cols)), dtype=object)

    for xi, x in enumerate(mat.T):
        for yi, y in enumerate(mat.T[xi:]):
            arr[xi, yi+xi] = list(map(lambda _: round(_,4), pearsonr(x,y)))
            arr[yi+xi, xi] = arr[xi, yi+xi]

    return pd.DataFrame(arr, index=cols, columns=cols)
```

```
corr_pvalue(df)
```

```
cat_names = ['cat0', 'cat1', 'cat2', 'cat3', 'cat4', 'cat5', 'cat6', 'cat7', 'cat8', 'cat9' ]
cont_names = ['cont0', 'cont1', 'cont2', 'cont3', 'cont4', 'cont5', 'cont6', 'cont7', 'cont8', 'cont9', 'cont10', 'cont11', 'cont12', 'cont13']
procs = [Categorify, Normalize]
```

```
x,y = int(0.2*df.shape[0]), int(0.25*df.shape[0])
```

```
splits = IndexSplitter(list(range(x,y)))(range_of(df))
```

```
to = TabularPandas(df, procs=procs, cat_names=cat_names, cont_names=cont_names, y_names="target",  splits=splits)
```

```
dls = to.dataloaders(bs=128)
```

```
dls.show_batch()
```

```
learn = tabular_learner(dls, layers=[200,100],metrics=rmse)
learn.fit(5, 1e-2)
```

```
learn.lr_find()
```

```
learn.fit_one_cycle(10, lr_max=slice(3e-4, 8e-4))
```

```
test_df = pd.read_csv(test_path)
test_df.head()
```

```
test_df.iloc[:,1:11] = test_df.iloc[:,1:11].apply(le.transform)
```
